```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def create_train_test_groups():
    train_groups = {
        0:[(0,1000000)],
        1:[(1500000,2000000),(3500000,4000000)],
        2:[(2500000,3000000),(4000000,4500000)],
        3:[(1000000,1500000),(3000000,3500000)],
        4:[(2000000,2500000),(4500000,5000000)]
    }
    train_signal_group = np.full(5000000, -1)
    for k,v in train_groups.items():
        for l,r in v:
            train_signal_group[l:r] = k


    test_signal_group = np.full(2000000, -1)
    test_groups = {
        0:[(0,100000)],
        1:[(100000,200000),(900000,1000000)],
        2:[(200000,300000),(600000,700000)],
        3:[(400000,500000)],
        4:[(500000,600000),(700000,800000)],
        5:[(300000,400000),(800000,900000),(1000000,2000000)]
    }
    for k,v in test_groups.items():
        for l,r in v:
            test_signal_group[l:r] = k
            
    return train_signal_group, test_signal_group

train_signal_group, test_signal_group = create_train_test_groups()
```

```python
train = np.load('train.npy')
oof_enc = np.load('oof_enc.npy')
oof_dec = np.load('oof_dec.npy')

test_enc = np.load('test_enc.npy')
test_dec = np.load('test_dec.npy')
```

```python
for i in np.random.randint(1125, size=20):
    plt.plot(train[i].reshape(-1))
    plt.plot(oof_dec[i].reshape(-1))
    plt.show()
```

```python
enc = np.concatenate([oof_enc, test_enc])
group = np.concatenate([train_signal_group[::4000],test_signal_group[::4000]])

np.unique(group)
```
